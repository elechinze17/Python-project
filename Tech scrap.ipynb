# Latest Funded Startups For 2023 Web scraping Project

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
url= 'https://growthlist.co/funded-startups/'

page = requests.get(url)

soup = BeautifulSoup(page.text, 'html')
```

```python
print(soup)
```

```python
soup.find_all('tr',class_ = 'footable-header')
```

```python
soup.find('tr',class_ = 'footable-header').text.strip()
```

```python
soup.find('tbody')
```

```python
table = soup.find_all('thead')
```

```python
print(table)
```

```python
world_tech = soup.find_all('th')
```

```python
world_table_tech = [ title.text.strip() for title in world_tech]

print(world_table_tech)
```

```python
import pandas as pd
```

```python
df = pd.DataFrame(columns = world_table_tech)

df
```

```python
col_data = soup.find_all('tr')
```

```python
for row in col_data[1:]:
   row_data = row.find_all('td', class_= "")
ind_row_data = [ data.text.strip() for data in row_data]

length = len(df)
df.loc[length] = ind_row_data
```
